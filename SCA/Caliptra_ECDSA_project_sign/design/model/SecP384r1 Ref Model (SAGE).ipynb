```sage
# Q=scalar*G over E
def ECPM(E, scalar, G, Q_expected):
    Q=scalar*G

    if (Q_expected):
        if (Q[0] == Q_expected[0]) & (Q[1] == Q_expected[1]):
            print('test passed!')
        else:
            print('test failed!')
    return Q
```
